```python
password = input('Please enter your password')
special_characters = ['!','@','#','$','%','^','&','*']
if not any(word.isupper() for word in password):
    print('Password must have uppercase')
elif not any(word.islower() for word in password):
    print('Password must have lowercase')
elif not any(word.isdigit() for word in password):
    print('Password must have at least a digit')
elif len(password) < 8:
    print('Password length must be a minimum of 8 characters')
elif not any(char in special_characters for char in password):
    print('Password must have at least 1 special character')
else:
    print('Password is valid')
```

```python
import random as rn
jrn = 'Junior'
itm = 'Intermediate'
srn = 'Senior'
li = []

counter = 1
while counter <= 500:
    n = rn.randint(100,2001)
    strn = str(n)
    
    if n >= 100 and n <= 1200:
        li.append(jrn + strn)
    elif n >= 1201 and n <= 1600:
        li.append(itm + strn)
    else:
        li.append(srn + strn)
        
        counter += 1
print(li)
```

```python
import random as rn
junior = 'JRN'
intermediate = 'ITM'
senior = 'SRN'
stafflist = []
counter = 1
while counter <= 500:
    num = rn.randint(100,2001)
    strnum =str(num)
    stafflist.append(junior + strnum)
    stafflist.append(intermediate + strnum)
    stafflist.append(senior + strnum)
    counter += 1
junior_staff = list(filter(lambda x: x.startswith('JRN'),stafflist))
print(junior_staff)
```
